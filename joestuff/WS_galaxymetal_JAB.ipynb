```python
#%%timeit

# First lets import in the data file: in this case '../data/galaxymetal_JAB.xlsx'
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)

# gala is a data table of a galaxymetal_JAB.xlsx file which 
# is subset of data from the galaxyinfo.xlsx data from CGM^2

gala = pd.read_excel('galaxymetal_JAB.xlsx')

gala.index += 2   # This is so the subset of data matches the exact index value of the galaxyinfo.xlsx data

gala.to_numpy()

gala

#gal = pd.ExcelFile('../data/galaxymetal_JAB.xlsx')
```

```python
# Now we will define a function to find distances to our galaxies.

def Distance(z):
    
    '''
    This function finds distance (units = cm) using 'FlatLambdaCDM' from astropy.cosmology
    
    z = redshift of galaxy
    '''
    
    cosmo = FlatLambdaCDM(H0 = 70*u.km/u.s/u.Mpc, Tcmb0 = 2.725*u.K, Om0 = 0.3)
    r = cosmo.luminosity_distance(z).to(u.cm)
    return r
```

```python
#%%timeit

# Now we find the distances to our galaxies and add a column to 'gala' with those distance values.

dist = Distance(gala['z'])
gala['Distance'] = dist

gala
```

```python
#%%timeit

# Now we're going to define some functions to finc the Luminosity of the galaxy using flux and distance
# Then we'll call the function on the H_alpha and H_beta flux from our subset of galaxy data.

def Luminosity_Ha(F, r):
    
    '''
    This function calculates the Luminosity of the galaxy given:
    
    F = H_alpha flux (in ergs/s)
    r = Distance to galaxy (in cm)
    '''
    
    LH_a = (F)*(4*np.pi)*(r)**2
        
    return LH_a

def Luminosity_Hb(F,r):
    
    '''
    This function calculates the Luminosity of the galaxy given:
    
    F = H_beta flux (in ergs/s)
    r = Distance to galaxy (in cm)
    '''
    
    LH_b = (2.86)*(F)*(4*np.pi)*(r)**2
    
    return LH_b


L_Ha = np.array(Luminosity_Ha(gala['Halpha_flux'], gala['Distance']))
L_Hb = np.array(Luminosity_Hb(gala['Hbeta_flux'], gala['Distance']))

#L_Ha
```

```python
#%%timeit

# Since we have '-99' values where there we didn't detect flux, we'll have to clean up the data where the 
# '-99' values were calculated. Since there will be negative values for flux (not possible) we'll turn
# all negatives values of Luminosity to '-99' to coincide with the rest of our dataset when there is 
# a null detection.

gala_1 = np.where(L_Ha < 0)
gala_2 = np.where(L_Hb < 0)
L_Ha[gala_1] = -99
L_Hb[gala_2] = -99
```

```python
#%%timeit

# Now we'll add 'Luminosity (ergs/s)' columns to our galaxy data subset dataframes.

gala['Luminosity_Ha'] = L_Ha
gala['Luminosity_Hb'] = L_Hb
gala
```

```python
# Defining a function to find SFR
# Try running some commands to slice down the data before computing.

def SFR(L):
    
    '''
    This function will calculate the star formation rates (SFR) using the Robert Kennicutt equation.
    
    L = Luminosity (ergs/s)
    '''
    
    RK = 7.9e-42
    SFR = ((RK)*(L))*((u.M_sun)/(u.yr))
    
    return SFR
```

```python
#%%timeit
# Here we find the SFR of the 'gala' dataframe and then adding the values to a new column.

SFR_Ha = np.array(SFR(gala['Luminosity_Ha']))
SFR_Hb = np.array(SFR(gala['Luminosity_Hb']))
```

```python
#%%timeit

# Again, we have to account for the '-99' values in our data.
# We then add the data to our 'gala' dataframe assigned to new columns.

gala_a = np.where(SFR_Ha < 0)
gala_b = np.where(SFR_Hb < 0)
SFR_Ha[gala_a] = -99
SFR_Hb[gala_b] = -99

gala['SFR_Ha'] = SFR_Ha
gala['SFR_Hb'] = SFR_Hb

gala
```

```python
#%%timeit

# Here we are just saving the dataframe 'gala' to a .csv file.

gala.to_csv('galaxy.csv', index=False)
galaxy = pd.read_csv('galaxy.csv')
galaxy
```

## Metallicity

```python
# Now we can define some functions to calculate the O3HB and N2 ratios for the galaxy data.

def metal_Hb(F_O, F_Hb):
    
    '''
    This function will calculate the flux ratio of Oxygen III (5007) and H_beta
    
    F_O = flux of oxygen III
    F_Hb = flux of H_beta
    '''
    
    O3Hb = F_O / F_Hb
    
    return O3Hb

def metal_Ha(F_N, F_Ha):
    
    '''
    This function will calculate the flux ratio of Nitrogen II (6583) and H_alpha
    
    F_N = flux of Nitrogen II
    F_Ha = Flux of H_alpha
    '''
    
    n2 = F_N / F_Ha
    
    return n2
```

```python
#%%timeit

# Now we'll use the function to find the O3HB values.
# Then we account for the '-99' values
# Finally we add a new column ('O3HB') to 'gala' 

O3HB = np.array(metal_Hb(gala['OIII_flux'],gala['Hbeta_flux']))
N2 = np.array(metal_Ha(gala['NII_flux'],gala['Halpha_flux']))


galax_1 = np.where((O3HB < 0) | (O3HB == 1.0))
O3HB[galax_1] = -99

galax_2 = np.where((N2 < 0) | (N2 == 1.0))
N2[galax_2] = -99

gala['O3HB'] = O3HB
gala['N2'] = N2

gala
```

```python
#%%timeit

# Here we'll take the log of the ratio of 'O3HB' and 'N2'
# We do some book keeping to account for the '-99' values.
# Finally we add a new column to our data.

ratio = np.array(gala['O3HB']/gala['N2'])

galax_a = np.where((ratio < 0) | (ratio == 1.0))  # This finds the ratios we don't wnat to use
ratio[galax_a] = 1           # This sets those unwanted ratios to a value of '1'

O3N2 = np.log10(ratio)
galax_b = np.where(O3N2 == 0.0)    # This finds all of the unwanted values (now 0.0) 
O3N2[galax_b] = -99                # This sets all of those unwated values to '-99' as per status quo
gala['O3N2'] = O3N2

gala
```

```python
# Now we'll use the Pettini & Pagel equations to calculate some Metallicity!

def metalow(O3N2_low):
    
    '''
    This function calculates metallicity (from Pettini and Pagel) for galaxies with O3N2 < 1.9
    
    O3N2_low = An array of O3N2 values.
    '''
    
    Z_l = 8.73-(0.32*O3N2_low)
    
    return Z_l
```

```python
#%%timeit

# Now we call the function to operate on the data from the 'gala' dataframe.
# Then we make all the values we don't want into '-99'
# And finally we add the data to a new column in the 'gala' dataframe.

Z_low = np.array(metalow(gala['O3N2']))

low_O3N2 = np.where((O3N2 < 0) | (O3N2 >= 1.9))
Z_low[low_O3N2] = -99

#Z_low_a = metalow(np.where((gala['O3N2'] > 0) | (gala['O3N2'] < 1.9)))

gala['Z_low'] = Z_low
gala

#gala['O3N2']
#low_O3N2
#np.where((gala['O3N2'] > 0) | (gala['O3N2'] < 1.9))
```

```python
def metalup(N2):
    
    '''
    This function calculates metallicity (from Pettini and Pagel) for galaxies with O3N2 > 1.9
    
    N2 = An array of N2 values
    '''
    
    Z_u = 8.92 + (0.57*np.log10(N2))
    
    return Z_u
```

```python
#%%timeit

# Here we are going to do something similar with values of O3N2 ≥ 1.9. There is only one in our current dataset.
# We also make a couple masks to find our values we don't want and brand them '-99'
# And of course we add another column to our 'gala' dataframe.

high_O3N2 = np.where((gala['O3N2'] > 1.9))
#print(high_O3N2[0])

no_N2 = np.where(gala['N2'] == -99)
act_Z_low = np.where(gala['Z_low'] != -99)
#print(len(no_N2[0]))

Z_up = np.array(metalup(gala['N2']))
Z_up[no_N2] = -99
Z_up[act_Z_low] = -99

gala['Z_up'] = Z_up
#print(Z_up[high_O3N2])
gala
```

```python
#%%timeit

# Let's get some R23 values.

def r23(OII,OIII,Hbeta):
    
    '''
    This is the R23 value used for calculating metallicity when there is no N2, but there are OII 
    and OII flux values.
    OII = OII flux
    OIII= OIII flux
    Hbeta= Hbeta flux
    '''
    
    r23 = (OII+OIII)/(Hbeta)
    
    return r23

O3HB = np.where((gala['O3HB'] == -99) | (gala['N2'] != -99))
no_O = np.where((gala['OII_LQ'] == -99) | (gala['OIII_LQ'] == -99))
no_Hb = np.where(gala['Hbeta_LQ'] == -99)

R23 = np.array(r23(gala['OII_flux'],gala['OIII_flux'],gala['Hbeta_flux']))
R23[O3HB] = -99
R23[no_O] = -99
R23[no_Hb]= -99

not_neg = np.where(R23 <= 0)
R23[not_neg] = -99
gala['R23'] = R23
gala
```

```python
def LO_UP(R23,OIII,OII):
    
    '''
    This function will bring in values of flux (R23,OIII,OII), define those values in an array
    of floats, and perform the necessary calculations to attain the upper (up) and lower (low) branch
    metallicity values using the equations given by McGaugh 1991. Note that this function does NOT
    decide which values are relevant for use with upper or lower branch depending on the degenracy
    of R23. This function performs the upper/lower branch calculations on ALL values for R23, upper
    and lower branch sorting will have to be performed after this function is called. This function
    returns two arrays, one of the lower branch, and one of the upper branch.
    
    R23 = Dataframe of R23 values that have previously been calculated
    OIII= Dataframe of OIII flux values
    OII = Dataframe of OII flux values
    '''
    
    r23 = np.array(R23, dtype=float)
    oiii= np.array(OIII, dtype=float)
    oii = np.array(OII, dtype=float)
    
    x = np.log10(r23)
    y = np.log10((1.3*(oiii))/oii)
    
    low = 12.0-4.944+0.767*x+0.602*x**2-y*(0.29+0.332*x-0.331*x**2)
    up  = 12.0-2.939-0.2*x-0.237*x**2-0.305*x**3-0.0283*x**4-y*(0.0047-0.0221*x-0.102*x**2-0.0817*x**3-0.00717*x**4)
    
    return np.array(low), np.array(up)
```

```python
#%%timeit

# Now we define two variables as arrays of our returns from the 'LO_UP' function

low_branch = LO_UP(gala['R23'],gala['OIII_flux'],gala['OII_flux'])[0]
up_branch  = LO_UP(gala['R23'],gala['OIII_flux'],gala['OII_flux'])[1]

# Now we will find where there wasn't a value for R23 (gala['R23'] == -99)and change our new arrays 
# (low_branch, up_branch) to reflect those null results.

no_r23 = np.where(gala['R23'] == -99)
low_branch[no_r23] = -99
up_branch[no_r23] = -99

```
